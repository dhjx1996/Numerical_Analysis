# Bisection Routine (Code from class with edits to sfprint)

```python
def bisection_method(a,b,f,tol,N):
    FA,FB = f(a), f(b)
    iteration = 1
    if FA*FB<0:
        print ("iter      a           b            root:(x,_)           root:(_,y)")
        print ("------------------------------------------------------------------")
        p = (a + b)/2
        FP = f(p)
        while (np.abs((b-a)/2)>tol) & (iteration <= N):
            if FA*FP<0:
                b=p
                FB=FP
            else:
                a=p
                FA=FP
            p = (a+b)/2
            FP = f(p)
            print('{:>3d}  {:> 10.5f}  {:> 10.5f}  {:> 22.16f}  {:>13.6e}'.format(iteration, a, b, p, abs(FP)))
            iteration = iteration + 1
        if np.abs((b-a)/2)>tol:
            print("Method failed to converge")
        else:
            return p
    else:
             print("Cannot ensure existence of root")
```

# Newton's Method (Code from class with edits to sfprint)

```python
def newton_method(x0,f,fp,tol,N):
    F = f(x0)
    Fp = fp(x0)
    iteration = 1
    print ("Original Newton's Method")
    print ("iter      grad         root:(x,_)           root:(_,y)")
    print ("------------------------------------------------------")
    while (iteration<=N) & (np.abs(F)>tol):
        x = (Fp*x0 - F)/Fp
        x0 = x
        F = f(x0)
        Fp = fp(x0)
        print('{:>3d}  {:> 10.5f}  {:> 22.16f}  {:>13.6e}'.format(iteration, Fp, x, abs(F)))
        iteration = iteration + 1
    if np.abs(F)<=tol:
        return x0
    else:
        print("Method failed to converge. Try harder!")
```

#### One possible proof for the divergence of the harmonic series: 

$\displaystyle p_n = \sum_{k=1}^n\frac{1}{k} > \int_0^n\frac{1}{x+1}\mathrm{d}x$ for all $n>1$.

As $n \rightarrow \infty$, $\displaystyle \int_0^n\frac{1}{x+1}\mathrm{d}x \rightarrow \infty$ which implies that the sequence $(p_n)$ diverges even though $p_n - p_{n-1}$ converges to 0.

Before we choose our stopping criteria we should ensure that it will be met when the number of iterations is sufficiently large. Regarding the specific stopping criterion $|p_n - p_{n-1}| < \epsilon$, it will be met if the function we are examining converges.

```python
def false_position_secant_method(x0,x1,f,tol,N):
    F0 = f(x0)
    F1 = f(x1)
    iteration = 2
    print("False Position Method")
    print ("iter       root:(x,_)           root:(_,y)")
    print ("------------------------------------------")
    while (iteration<=N) & (np.abs(F1)>tol):
        x2 = x1 - F1*(x1-x0)/(F1-F0)
        if f(x2)*f(x1) < 0:
            x0 = x1
            x1 = x2
        else:
            x0 = x0
            x1 = x2
        F0 = f(x0)
        F1 = f(x1)
        print('{:>3d}  {:> 22.16f}  {:>13.6e}'.format(iteration, x2.real, abs(F1.real)))
        iteration = iteration + 1
    if np.abs(F1)<=tol:
        return x2
    else:
        print("Method failed to converge. Try harder!")
        
def secant_method(x0,x1,f,tol,N):
    F0 = f(x0)
    F1 = f(x1)
    iteration = 2
    print("Secant Method")
    print ("iter       root:(x,_)           root:(_,y)")
    print ("------------------------------------------")
    while (iteration<=N) & (np.abs(F1)>tol):
        x = x1 - F1*(x1-x0)/(F1-F0)
        x0 = x1
        x1 = x
        F0 = f(x0)
        F1 = f(x1)
        print('{:>3d}  {:> 22.16f}  {:>13.6e}'.format(iteration, x.real, abs(F1.real)))
        iteration = iteration + 1
    if np.abs(F1)<=tol:
        return x1
    else:
        print("Method failed to converge. Try harder!")
```
